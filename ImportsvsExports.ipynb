```python
# Comex Stat Import Data (Brazil) vs UN Comtrade Export Data (France)

#This notebook compares Brazilian foie gras import records from Comex Stat with French export records from UN Comtrade.

#Goal

#Identify discrepancies between the two databases, focusing on differences in reported unit values (US$/kg).

## Notebook structure

#Data loading and exploration
#Unit value comparison
#Database analysis
#Results and interpretation

import pandas as pd

comex = pd.read_csv("comexstat.csv")
comtrade = pd.read_csv("comtrade.csv")
```

```python
comex = comex.rename(columns={"Price US$/kg": "Comex_Price_USD_kg"})
comtrade = comtrade.rename(columns={"Price US$/kg": "Comtrade_Price_USD_kg"})
```

```python
comex = comex[["Year", "Comex_Price_USD_kg"]]
comtrade = comtrade[["Year", "Comtrade_Price_USD_kg"]]
```

```python
comparison = pd.merge(
    comex,
    comtrade,
    on="Year"
)
```

```python
comparison["Difference_USD_kg"] = (
    comparison["Comtrade_Price_USD_kg"] -
    comparison["Comex_Price_USD_kg"]
)
```

```python
largest_difference = comparison.loc[
    comparison["Difference_USD_kg"].idxmax()
]
```

```python
print(largest_difference)
```

```python
print (comparison)
```

```python
comparison["Difference_percent"] = (
    (comparison["Comtrade_Price_USD_kg"] - comparison["Comex_Price_USD_kg"])
    / comparison["Comex_Price_USD_kg"]
) * 100

```
