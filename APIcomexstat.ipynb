```python


payload = {
    "flow": "import",
    "monthDetail": True,
    "period": {
        "from": "2019-01",
        "to": "2026-12"
    },
    "filters": [
        {
            "filter": "country",
            "values": [275]   # França
        },
        {
            "filter": "ncm",
            "values": [
                "02074300",
                "02075300"
            ]
        }
    ],
    "details": [
        "country",
        "ncm"
    ],
    "metrics": [
        "metricFOB",
        "metricKG",
        "metricCIF",
        "metricStatistic"
    ]
}

r = requests.post(url, json=payload)

print(r.status_code)
print(r.json())
```
