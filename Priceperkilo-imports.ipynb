```python
# Foie Gras Import Prices per Kilogram (Brazil)

#This notebook analyzes Brazilian foie gras import data to calculate changes in average price per kilogram over time.

# Goal

#Identify trends in foie gras import prices and determine the highest unit value recorded during the period analyzed.

# Notebook structure

#Data loading and exploration
#Calculation of price per kilogram
#Results and interpretation

import pandas as pd
```

```python
df = pd.read_csv("comexstatimports.csv")
```

```python
df.head(8)
```

```python
lowest_price = df.loc[df["Price US$/kg"].idxmin()]
```

```python
highest_price = df.loc[df["Price US$/kg"].idxmax()]
```

```python
increase = (
    (highest_price["Price US$/kg"] - lowest_price["Price US$/kg"])
    / lowest_price["Price US$/kg"]
) * 100
```
